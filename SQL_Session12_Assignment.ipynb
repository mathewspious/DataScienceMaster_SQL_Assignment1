```python
## Importing packages
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import sqlite3 as db
```

```python
## Reading the dataset from url
url='https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
## Assigning the Dataset to DF
census_data=pd.read_csv(url,names=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','XYX'])
```

```python
## Verifiying the dataset
census_data.head()
```

```python
## Dataset Info
census_data.info()
```

```python
## Creating DB
conn=db.connect('sqladb.db')
```

```python
## Creating Table
with conn:
    cur = conn.cursor() 
    cur.execute("CREATE TABLE adult(age INT ,workclass TEXT ,fnlwgt INT ,education TEXT ,'education-num' INT  ,'marital-status' TEXT ,occupation TEXT ,relationship TEXT ,race TEXT ,sex TEXT ,'capital-gain' INT ,'capital-loss' INT ,hours INT ,'native-country' TEXT ,XYX TEXT)")
```

```python
## Writing data from Dataframe to DB
census_data.to_sql('adult', conn, if_exists='replace')
```

#### 1. Select 10 records from adult sqladb

```python
 # A SQL command to get 10 rows
ten_record_sql = "SELECT * FROM adult LIMIT 10;" 
```

```python
for result in cur.execute(ten_record_sql):
    print(result)
```

#### 2. Avg hours per week for all en working in Private sector

```python
private_workhour_sql = "SELECT AVG(\"hours-per-week\") FROM adult WHERE workclass=' Private' AND sex=' Male' GROUP BY sex;"
for result in cur.execute(private_workhour_sql):
    print(result)
```

#### 3. Frequncy Tables 

#### 3A. Frequency Table :-  Education

```python
edu_freq_sql = "SELECT education,COUNT(*) FROM adult GROUP BY education;" #SHOW EDUCATION TABLE
for result in cur.execute(edu_freq_sql):
    print(result)
```

#### 3B. Frequenct Table :- Occupation

```python
occupation_freq_sql = "SELECT occupation,COUNT(*) FROM adult GROUP BY occupation;" #SHOW occupation TABLE
for result in cur.execute(occupation_freq_sql):
    print(result)
```

#### 3C. Frequncy Table Relationship

```python
relationship_freq_sql = "SELECT relationship,COUNT(*) FROM adult GROUP BY relationship;" #SHOW Relationship TABLE
for result in cur.execute(relationship_freq_sql):
    print(result)
```

#### 4. People who are married, working in private sector and having masters degree

```python
select_sql = "SELECT count(*) FROM adult WHERE workclass=' Private' AND education=' Masters' AND \"marital-status\" != ' Never-married';"
for result in cur.execute(select_sql):
    print(result)
```

#### 5. Avg, Min and Max age Group of people working inn different Sectors

```python
age_group_sql = "SELECT workclass,MAX(age),MIN(age),AVG(age) FROM adult GROUP BY workclass;"
for result in cur.execute(age_group_sql):
    print(result)
```

#### 6. Age distribution by country

```python
age_dist_country_sql = "SELECT \"native-country\",age,COUNT(*) FROM adult GROUP BY \"native-country\",age;"
for result in cur.execute(age_dist_country_sql):
    print(result)
```
